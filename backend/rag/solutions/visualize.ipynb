```python
import os
import matplotlib.pyplot as plt
```

```python
basic_info = []
trinhtuthuchien = []
cachthucthuchien = []
hoso = []
coquan = []
cancuphaply = []
count = 0
for filename in os.listdir("../../crawler/data/thutuchanhchinh_preprocessed/"):
    with open("../../crawler/data/thutuchanhchinh_preprocessed/" + filename, "r") as f:
        text = f.read()
        texts = text.split("\n\n")
        if len(texts) != 7:
            count += 1
            continue
        basic_info.append(texts[0])
        trinhtuthuchien.append(texts[1])
        cachthucthuchien.append(texts[2])
        hoso.append(texts[3])
        coquan.append(texts[4])
        cancuphaply.append(texts[5])
        f.close()
```

```python
# Draw histogram for character length of basic_info
plt.hist([len(x) for x in basic_info], range=(300, 1200))
plt.title("Biểu đồ phân hóa số lượng ký tự thông tin chung")
plt.xlabel("Character length")
plt.ylabel("Frequency")
plt.show()
```

```python
# Draw histogram for character length of trinhtuthuchien
plt.hist([len(x) for x in trinhtuthuchien], range=(0,8000))
plt.title("Biểu đồ phân hóa số lượng ký tự của trình tự thực hiện")
plt.xlabel("Số lượng ký tự")
plt.ylabel("Số đoạn")
plt.show()
```

```python
# Draw histogram for character length of cachthucthuchien
plt.hist([len(x) for x in cachthucthuchien], range=(300, 2500))
plt.title("Biểu đồ phân hóa số lượng ký tự của cách thức thực hiện")
plt.xlabel("Số lượng ký tự")
plt.ylabel("Số đoạn")
plt.show()
```

```python
# Draw histogram for character length of trinhtuthuchien
plt.hist([len(x) for x in hoso], range=(0, 5200))
plt.title("Biểu đồ phân hóa số lượng ký tự của thành phần hồ sơ")
plt.xlabel("Số lượng ký tự")
plt.ylabel("Số đoạn")
plt.show()
```

```python
# Draw histogram for character length of coquan
plt.hist([len(x) for x in coquan], range=(0, 5000))
plt.title("Biểu đồ phân hóa số lượng ký tự của thông tin cơ quan")
plt.xlabel("Số lượng ký tự")
plt.ylabel("Số đoạn")
plt.show()
```

```python
# Draw histogram for character length of cancuphaplys
plt.hist([len(x) for x in cancuphaply], range=(0, 2000))
plt.title("Biểu đồ phân hóa số lượng ký tự của căn cứ pháp lý")
plt.xlabel("Số lượng ký tự")
plt.ylabel("Số đoạn")
plt.show()
```
